## Task 9: Mini use case (Sales data analysis)

```python
sales={
    'Day':['Mon','Tue','Wed','Thu','Fri'],
    'Revenue':[1200,1500,900,2000,1800]
}
```

```python
sales_df=pd.DataFrame(sales)
```

```python
sales_df
```

```python
print("Total Revenue: ",sales_df['Revenue'].sum())
```

```python
print("Avarage Revenue: ",sales_df['Revenue'].mean())
```

```python
print("Highest Revenue: \n",sales_df[sales_df['Revenue']==sales_df['Revenue'].max()])
```

```python
sales_df[sales_df['Revenue']>sales_df['Revenue'].mean()]
```

```python
sales_df.plot(kind='bar',y='Revenue',x='Day')
```
