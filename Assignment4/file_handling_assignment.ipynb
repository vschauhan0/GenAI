# Assignment 4: File Handling

## Task 1 – Write Sales Records to a File

```python
sales = [1200, 450, 980, 1500, 3000]

# Write data to file
with open("sales_data.txt", "w") as file:
    for sale in sales:
        file.write(str(sale) + "\n")

# Reopen and print content
with open("sales_data.txt", "r") as file:
    print(file.read())
```

## Task 2 – Read File in Different Ways

```python
# Read entire file
with open("sales_data.txt", "r") as file:
    print("Using read():")
    print(file.read())

# Read first line
with open("sales_data.txt", "r") as file:
    print("\nUsing readline():")
    print(file.readline())

# Read all lines and convert to integers
with open("sales_data.txt", "r") as file:
    lines = file.readlines()
    sales_list = [int(line.strip()) for line in lines]

print("\nConverted List:", sales_list)
```

## Task 3 – Append New Sales

```python
new_sales = [5000, 2500, 1700]

# Append new sales
with open("sales_data.txt", "a") as file:
    for sale in new_sales:
        file.write(str(sale) + "\n")

# Read updated file
with open("sales_data.txt", "r") as file:
    lines = file.readlines()
    print("Updated File:")
    print("".join(lines))

print("Total number of lines:", len(lines))
```

## Task 4 – Generate Summary Report

```python
with open("sales_data.txt", "r") as file:
    sales = [int(line.strip()) for line in file.readlines()]

total_sales = sum(sales)
highest_sale = max(sales)
lowest_sale = min(sales)
average_sale = total_sales / len(sales)

print("Total Sales:", total_sales)
print("Highest Sale:", highest_sale)
print("Lowest Sale:", lowest_sale)
print("Average Sale:", average_sale)
```

## Task 5 – Create Product Info File (User Input)

```python
products = []

n = int(input("How many products? "))

for i in range(n):
    name = input("Enter product name: ")
    price = input("Enter price: ")
    products.append(name + " | " + price)

# Write to file
with open("products.txt", "w") as file:
    for item in products:
        file.write(item + "\n")

# Read and print properly
with open("products.txt", "r") as file:
    print("\nProduct List:")
    for line in file:
        name, price = line.strip().split("|")
        print("Product:", name.strip(), " Price:", price.strip())
```

## Task 6 – Read File Safely

```python
import os

filename = input("Enter filename to open: ")

if os.path.exists(filename):
    with open(filename, "r") as file:
        print(file.read())
else:
    print("File not found. Please check the filename.")
```

## Task 7 – Mini Project: Export Discounted Prices

```python
prices = {
    "Mouse": 500,
    "Keyboard": 800,
    "Monitor": 7000,
    "Pendrive": 400,
    "Camera": 5000
}

discount = float(input("Enter discount percentage: "))

discounted_prices = {}

# Write discounted prices
with open("discount_report.txt", "w") as file:

    total = 0
    count = 0

    for product, price in prices.items():
        discounted_price = price - (price * discount / 100)
        discounted_prices[product] = discounted_price

        file.write(product + " | " + str(price) + " | " + str(discounted_price) + "\n")

        total += discounted_price
        count += 1

    # Extra summary
    avg = total / count
    file.write("\nTotal Items: " + str(count))
    file.write("\nAverage Discounted Price: " + str(avg))

# Read and print file
with open("discount_report.txt", "r") as file:
    print(file.read())
```
