## Task 3: Product Pricing

```python
price_dict = {
    "Laptop": 65000,
    "Smartphone": 30000,
    "Headphones": 2500,
    "Keyboard": 1500,
    "Mouse": 800,
    "Monitor": 12000
}
```

```python
price_dict["Webcam"] = 3500

# Update price of existing product
price_dict["Mouse"] = 900
```

```python
product_to_remove = "Printer"
removed_price = price_dict.pop(product_to_remove, None)
if removed_price is None:
    print(f"{product_to_remove} not found in price list.")
```

```python
average_price = sum(price_dict.values()) / len(price_dict)
print("Average product price:", average_price)
```

## Task 4: Combined Operations

```python
catalog = []
for i in range(len(products)):
    product_name = products[i]
    price = price_dict.get(product_name, 0)
    category = categories[i]
    catalog.append((product_name, price, category))

print("Catalog:", catalog)
```

```python
category_to_products = {}

for product, price, category in catalog:
    category_to_products.setdefault(category, []).append(product)

print("Category to products:", category_to_products)
```

```python
max_category = max(category_to_products, key=lambda c: len(category_to_products[c]))
print("Category with maximum products:", max_category)
print("Products in this category:", category_to_products[max_category])
```
