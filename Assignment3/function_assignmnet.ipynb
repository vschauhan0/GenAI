## Task 1 – Basic Function: Price After Discount

```python
def apply_discount(price, discount_percent=5):

    # Ensure discount never exceeds 60%
    if discount_percent > 60:
        discount_percent = 60

    final_price = price - (price * discount_percent / 100)
    return final_price


# Testing
print(apply_discount(1000, 10))
print(apply_discount(500))   # uses default discount
```

## Task 2 – Recursive Function: Factorial Utility

```python
def factorial(n):

    if n < 0:
        print("Error: Negative number not allowed")
        return None

    if n == 0 or n == 1:
        return 1

    return n * factorial(n - 1)


# Testing
print(factorial(5))
print(factorial(0))
print(factorial(-3))
```

## Task 3 – Lambda Function: GST Calculator

```python
gst = lambda price: price + (0.18 * price)

print(gst(100))


# Optional → GST + Discount together
gst_discount = lambda price, discount: (price + 0.18 * price) * (1 - discount / 100)

print(gst_discount(100, 10))
```

## Task 4 – Using map(): Apply GST to List

```python
prices = [100, 250, 400, 1200, 50]

gst = lambda price: price + (0.18 * price)

prices_with_gst = list(map(gst, prices))

print("Original Prices:", prices)
print("Prices after GST:", prices_with_gst)
```

## Task 5 – Using filter(): Filter Expensive Products

```python
prices = [100, 250, 400, 1200, 50, 2000, 850]

above_500 = list(filter(lambda x: x > 500, prices))
below_equal_500 = list(filter(lambda x: x <= 500, prices))

print("Prices > 500:", above_500)
print("Prices <= 500:", below_equal_500)
```

## Task 6 – Combined Utility Function

```python
def process_prices(prices):

    # Apply 10% discount using map
    discounted_prices = list(map(lambda x: x * 0.9, prices))

    # Keep only prices above 300 using filter
    filtered_prices = list(filter(lambda x: x > 300, discounted_prices))

    return discounted_prices, filtered_prices


# Testing
print(process_prices([100, 500, 900, 50, 750]))
```

## Task 7 – Mini Problem: Menu Using Functions

```python
def add_price(prices_list, price):
    prices_list.append(price)

def get_average_price(prices_list):
    if len(prices_list) == 0:
        return 0
    return sum(prices_list) / len(prices_list)

def get_max_price(prices_list):
    if len(prices_list) == 0:
        return 0
    return max(prices_list)


prices = []

while True:

    print("\nMenu")
    print("1 - Add price")
    print("2 - Show average price")
    print("3 - Show highest price")
    print("q - Quit")

    choice = input("Enter choice: ")

    if choice == "1":
        price = float(input("Enter price: "))
        add_price(prices, price)

    elif choice == "2":
        print("Average Price:", get_average_price(prices))

    elif choice == "3":
        print("Highest Price:", get_max_price(prices))

    elif choice == "q":
        break

    else:
        print("Invalid choice")
```
